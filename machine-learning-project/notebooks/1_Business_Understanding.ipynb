### **Contexto:** 

EA Sports FIFA, conocido simplemente como FIFA, es una saga de videojuegos de fútbol publicada anualmente por Electronic Arts bajo el sello de EA Sports, en colaboración con la FIFA.

El juego busca replicar la experiencia del fútbol real, tanto en la gestión de equipos como en la competencia dentro del campo, ofreciendo realismo gracias a sus licencias oficiales de equipos, jugadores y ligas de todo el mundo. Con la reciente transición a EA Sports FC, el objetivo principal sigue siendo simular el deporte del fútbol y permitir a los jugadores disputar partidos o gestionar un club en diversos modos de juego.

Se utilizó BeautifulSoup para extraer datos desde Sofifa.com
. Primero, se procesó la página principal para obtener información general de los jugadores y luego se accedió a la página individual de cada jugador para capturar datos específicos. Para optimizar la extracción, se definió un tamaño de lote que permitió paralelizar la recuperación de información. Posteriormente, se fusionaron todos los DataFrames y se realizó la limpieza del conjunto de datos final.

El dataset original fue publicado por Bryan B. en Kaggle
 y abarca información de futbolistas profesionales desde 2017 hasta 2023. Contiene más de 17 000 registros con atributos demográficos, características físicas, estadísticas de juego, detalles contractuales y afiliaciones a clubes.

### **Columnas:**

- **ID:** Un identificador único para cada jugador.
- **Name** El nombre del jugador.
- **Age:** La edad del jugador en el momento de la recopilación de datos.
- **Photo:** Un enlace o referencia a la fotografía del jugador.
- **Nationality:** La nacionalidad del jugador.
- **Flag:** La bandera nacional asociada a la nacionalidad del jugador.
- **Overall:** La calificación general de las habilidades y capacidades del jugador.
- **Potential:** La calificación potencial que representa el desarrollo futuro del jugador.
- **Club:** La afiliación actual del club del jugador.
- **Club Logo:** Un enlace o referencia al logotipo del club del jugador.
- **Value(£):** El valor de mercado estimado del jugador en libras (£).
- **Wage(£):** El salario semanal del jugador en libras (£).
- **Special:** Un valor numérico que representa las habilidades especiales del jugador.
- **Preferred Foot:** el pie preferido del jugador para jugar.
- **International Reputation:** una calificación que indica la reputación internacional del jugador.
- **Weak Foot:** una calificación que representa las habilidades del pie más débil del jugador.
- **Skill Moves:** La cantidad de movimientos de habilidad que posee el jugador.
- **Work Rate:** El ritmo de trabajo del jugador.
- **Body Type:** La constitución física o tipo de cuerpo del jugador.
- **Real Face:** indica si el jugador tiene una representación de cara real.
- **Position:** La posición de juego preferida del jugador.
- **Joined:** La fecha en que el jugador se unió al club actual.
- **Loaned From:** El club en el que el jugador se encuentra actualmente cedido.
- **Contract Valid Until:** La fecha hasta la cual el contrato del jugador es válido.
- **Height(cm.):** La altura del jugador en centímetros.
- **Weight(lbs.):** El peso del jugador en libras.
- **Release Clause(£):** El valor de la cláusula de liberación del jugador en libras (£).
- **Kit Number:** El número de kit del jugador. (no existe)
- **Best Overall Rating:** La calificación general más alta del jugador.
- **Year_Joined:** El año en que el jugador se unió al club actual.
