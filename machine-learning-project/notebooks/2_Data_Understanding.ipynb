### Diagramas de caja
Muestra cómo se reparte el sueldo de los jugadores según la edad.
Cada caja es un rango de edad: la línea del medio es el sueldo “típico” y la altura de la caja indica qué tan dispersos están los sueldos en ese grupo. Al usar escala log, se ven mejor tanto sueldos bajos como muy altos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- helper: asegurar Wage_num y AgeBin en cada df ---
def prepare_wage_agebins(df, bins=None, labels=None):
    df = df.copy()
    if bins is None:
        bins = [16, 20, 25, 30, 35, 40]
    if labels is None:
        labels = ['16-20', '21-25', '26-30', '31-35', '36+']
    # Wage -> numérico
    if "Wage_num" not in df.columns:
        df['Wage_num'] = (df['Wage']
                          .astype(str)
                          .str.replace('€','', regex=False)
                          .str.replace('K','e3', regex=False)
                          .str.replace('M','e6', regex=False))
        df['Wage_num'] = pd.to_numeric(df['Wage_num'], errors='coerce')
    # Age bins
    if "AgeBin" not in df.columns:
        df['AgeBin'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
    return df, labels

# Prepara cada año
FIFA20_df_prep, labels = prepare_wage_agebins(FIFA20_df)
FIFA21_df_prep, _      = prepare_wage_agebins(FIFA21_df, labels=labels)
FIFA22_df_prep, _      = prepare_wage_agebins(FIFA22_df, labels=labels)

# Arma datos para boxplot por año
def box_data_by_agebin(df, labels):
    sub = df.dropna(subset=['Wage_num','AgeBin'])
    return [sub.loc[sub['AgeBin']==b, 'Wage_num'].values for b in labels]

data20 = box_data_by_agebin(FIFA20_df_prep, labels)
data21 = box_data_by_agebin(FIFA21_df_prep, labels)
data22 = box_data_by_agebin(FIFA22_df_prep, labels)

# --- Figura con 3 subplots (2020, 2021, 2022) ---
fig, axes = plt.subplots(1, 3, figsize=(16, 5), sharey=True)

for ax, data, year in zip(axes, [data20, data21, data22], [2020, 2021, 2022]):
    ax.boxplot(data, showfliers=False)
    ax.set_yscale('log')
    ax.set_xticks(range(1, len(labels)+1))
    ax.set_xticklabels(labels, rotation=0)
    ax.set_title(f"Salario (€) por rango de edad — {year}")
    ax.set_xlabel("Rango de edad")
    ax.grid(axis='y', alpha=0.2)

axes[0].set_ylabel("Wage (escala log)")
plt.tight_layout()
plt.show()

# ====== (Opcional) Comparativa de MEDIANAS por rango de edad ======
def medians_by_agebin(df, labels):
    sub = df.dropna(subset=['Wage_num','AgeBin'])
    med = sub.groupby('AgeBin')['Wage_num'].median().reindex(labels)
    return med.values

med20 = medians_by_agebin(FIFA20_df_prep, labels)
med21 = medians_by_agebin(FIFA21_df_prep, labels)
med22 = medians_by_agebin(FIFA22_df_prep, labels)

plt.figure(figsize=(8,5))
x = np.arange(len(labels))
plt.plot(x, med20, marker='o', label='2020')
plt.plot(x, med21, marker='o', label='2021')
plt.plot(x, med22, marker='o', label='2022')
plt.yscale('log')
plt.xticks(x, labels)
plt.ylabel("Mediana de Wage (log)")
plt.xlabel("Rango de edad")
plt.title("Comparación de salarios medianos por edad (2020–2022)")
plt.legend()
plt.tight_layout()
plt.show()
```

Arriba (3 boxplots, uno por año): compara cómo se reparten los sueldos en cada rango de edad para 2020, 2021 y 2022.

La línea del medio es el sueldo típico (mediana).

Cajas más altas ⇒ sueldos más altos; cajas más grandes ⇒ más diferencia entre jugadores de ese rango.

Con escala log ves bien tanto sueldos bajos como muy altos.

Abajo (líneas de medianas): resume la mediana de sueldo por rango de edad y la compara entre años.

Lo usual es que los sueldos suban en 21–30, se estabilicen o piquen en 31–35 y luego bajen en 36+.

Si una línea (por ejemplo 2022) queda más arriba, ese año pagó mejor en ese rango; si las líneas se acercan o separan, cambió la brecha entre edades.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ===== Configuración común =====
AGE_BINS   = [15, 20, 25, 30, 35, 45]
AGE_LABELS = ["16-20","21-25","26-30","31-35","36+"]

def prepare_growthgap(df):
    """Asegura numéricos, calcula GrowthGap y arma AgeBin compatible."""
    x = df.copy()
    for c in ["Overall","Potential","Age"]:
        x[c] = pd.to_numeric(x.get(c), errors="coerce")
    x["GrowthGap"] = x["Potential"] - x["Overall"]
    x["AgeBin"] = pd.cut(
        x["Age"], bins=AGE_BINS, labels=AGE_LABELS,
        include_lowest=True, right=True, ordered=True
    ).cat.set_categories(AGE_LABELS, ordered=True)
    return x

def box_data_by_agebin(df):
    sub = df.dropna(subset=["GrowthGap","AgeBin"])
    return [sub.loc[sub["AgeBin"]==b, "GrowthGap"].values for b in AGE_LABELS]

# ===== Prepara cada año (usa tus dataframes FIFA20_df/FIFA21_df/FIFA22_df) =====
g20 = prepare_growthgap(FIFA20_df)
g21 = prepare_growthgap(FIFA21_df)
g22 = prepare_growthgap(FIFA22_df)

data20 = box_data_by_agebin(g20)
data21 = box_data_by_agebin(g21)
data22 = box_data_by_agebin(g22)

# ===== Boxplots lado a lado =====
fig, axes = plt.subplots(1, 3, figsize=(16, 5), sharey=True)

for ax, data, year in zip(axes, [data20, data21, data22], [2020, 2021, 2022]):
    ax.boxplot(data, showfliers=False)
    ax.set_xticks(range(1, len(AGE_LABELS)+1))
    ax.set_xticklabels(AGE_LABELS)
    ax.set_title(f"Brecha de crecimiento por edad — {year}")
    ax.set_xlabel("Rango de edad")
    ax.grid(axis='y', alpha=0.2)

axes[0].set_ylabel("GrowthGap (Potential − Overall)")
plt.tight_layout()
plt.show()
```

```python
# Calcula y muestra las medianas para verificar
med20 = (g20.dropna(subset=["GrowthGap","AgeBin"])
           .groupby("AgeBin", observed=True)["GrowthGap"]
           .median()
           .reindex(AGE_LABELS))
med21 = (g21.dropna(subset=["GrowthGap","AgeBin"])
           .groupby("AgeBin", observed=True)["GrowthGap"]
           .median()
           .reindex(AGE_LABELS))
med22 = (g22.dropna(subset=["GrowthGap","AgeBin"])
           .groupby("AgeBin", observed=True)["GrowthGap"]
           .median()
           .reindex(AGE_LABELS))

print("med20:\n", med20)
print("med21:\n", med21)
print("med22:\n", med22)

# Plot con estilos/offset para evitar superposición
import numpy as np
x = np.arange(len(AGE_LABELS))

plt.figure(figsize=(8,5))
plt.plot(x-0.04, med20.values, marker='o', linestyle='-',  linewidth=2, label='2020')
plt.plot(x+0.00, med21.values, marker='s', linestyle='--', linewidth=2, label='2021')
plt.plot(x+0.04, med22.values, marker='^', linestyle='-.', linewidth=2, label='2022')

plt.xticks(x, AGE_LABELS)
plt.ylabel("Mediana de GrowthGap")
plt.xlabel("Rango de edad")
plt.title("Comparación de brecha de crecimiento por edad (2020–2022)")
plt.legend()
plt.tight_layout()
plt.show() 
```

El gráfico muestra que los jugadores jóvenes (16–20) tienen la mayor brecha de crecimiento y que esta disminuye con la edad, casi desapareciendo en los mayores de 30. La tendencia se repite en 2020, 2021 y 2022, lo que indica un patrón estable: a menor edad, mayor potencial de mejora.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- helpers ---
def parse_euro_money(s):
    if pd.isna(s): return np.nan
    s = str(s).replace('€','').replace(',','').replace('K','e3').replace('M','e6')
    return pd.to_numeric(s, errors='coerce')

def prep_wage(df):
    """Normaliza columnas y agrega Wage_num si no existe."""
    x = df.copy()
    x.columns = x.columns.str.strip().str.replace(' ', '_')
    if 'Wage_num' not in x.columns:
        x['Wage_num'] = x['Wage'].map(parse_euro_money)
    return x.dropna(subset=['Wage_num','Club'])

def top_clubs(df, by='mean', top=12):
    if by == 'median':
        s = df.groupby('Club')['Wage_num'].median().sort_values(ascending=False)
    else:
        s = df.groupby('Club')['Wage_num'].mean().sort_values(ascending=False)
    return s.head(top).index.tolist()

def build_box_data(df, clubs_order):
    """Devuelve lista de arrays de salarios en el orden de clubs_order."""
    return [df.loc[df['Club']==c, 'Wage_num'].values for c in clubs_order]

# --- configura qué comparar ---
dfs   = [FIFA20_df, FIFA21_df, FIFA22_df]
years = [2020, 2021, 2022]
TOP_N = 12
ORDER_BY = 'mean'   # 'mean' o 'median'

# --- prepara dataframes por año ---
ready = [prep_wage(d) for d in dfs]

# --- obtiene tops por año y decide conjunto común ---
tops_per_year = [top_clubs(d, by=ORDER_BY, top=TOP_N) for d in ready]
common = set(tops_per_year[0]).intersection(*tops_per_year[1:])

if len(common) >= max(6, TOP_N // 2):
    clubs = sorted(common)          # mismos clubes en los 3 años
    subtitle = f"(mismos clubes en los 3 años: {len(clubs)})"
else:
    # si la intersección es chica, usa la unión y limita a TOP_N por alguna métrica global
    union = list(set().union(*tops_per_year))
    # orden global por salario medio 2022 (puedes cambiar a otra referencia)
    ref_rank = (ready[-1].groupby('Club')['Wage_num']
                .mean().sort_values(ascending=False))
    clubs = [c for c in ref_rank.index if c in union][:TOP_N]
    subtitle = f"(conjunto combinado, ordenado por {ORDER_BY} {years[-1]})"

# --- arma datos para los boxplots en el mismo orden de clubes ---
box_by_year = [build_box_data(d, clubs) for d in ready]

# --- plot: 3 subplots lado a lado, misma escala y orden de clubes ---
fig, axes = plt.subplots(1, 3, figsize=(16, 7), sharey=True)

for ax, data, yr in zip(axes, box_by_year, years):
    ax.boxplot(data, showfliers=False, vert=False)  # horizontal para leer mejor
    ax.set_xscale('log')
    ax.set_yticks(range(1, len(clubs)+1))
    ax.set_yticklabels(clubs)
    ax.set_xlabel('Wage (escala log)')
    ax.set_title(f"Top {TOP_N} por {ORDER_BY} — {yr}")

axes[0].set_ylabel('Club')
plt.suptitle(f"Distribución de salarios por club — comparación {years[0]}–{years[-1]} {subtitle}", y=0.98)
plt.tight_layout()
plt.show()
```

El gráfico muestra cómo se distribuyen los salarios en los clubes con mayores sueldos entre 2020 y 2022. Se observa que algunos equipos se mantienen consistentemente en lo más alto del ranking, pagando salarios mayores año tras año. La mediana de sueldos permite ver el nivel típico de cada club, mientras que la altura de las cajas refleja cuán pareja o desigual es la distribución dentro del plantel. En general, los clubes top combinan altos salarios promedio con brechas internas marcadas, aunque en algunos casos se aprecian ajustes entre un año y otro.
