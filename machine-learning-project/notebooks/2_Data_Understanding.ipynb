### Naturaleza de los datos

En el contexto de los datos provenientes del videojuego FIFA, la clasificación de las variables se realiza considerando su representación dentro del juego y su uso potencial en análisis estadísticos o modelos de machine learning.

Ejemplo Penalties: Cuando hablamos de Penalties no hacemos referencia a cuantos penales efectua o realiza un jugador como si fuera una variable discreta en la cual son numeros enteros, sino, que estamos midiendo una habilidad o en su efectividad en esa acción. Siendo un valor calculado o medido. Para este ejemlo se toma de referencia para las demas variables.

Tipo de dato: Como el numero de la camiseta del jugador no se puede calcular, entonces es una variable numerica discreta aun que su formato este en flotante.

Variables que contienen links o IDs no seran tomado en cuenta para realizar modelos de machine learning.

```python
# Variables Categoricas ordinales y nominales


#N00000 existe relaciónde orden en suscategorías.
var_categoricas_nominales = ['Name','Nationality','Club','Preferred Foot',
                            'Body Type','Real Face','Position','Loaned From',
                            'Best Position']

#Existe relación de orden en sus categorías.
var_categoricas_ordinales = ['Work Rate']

# Variables numericas discretas y continuas

# Numeros enteros: 1,2,3
var_numericas_discretas = ['Age','Overall','Special','Skill Moves',
                          'Jersey Number', 'Year_Joined','Best Overall Rating',
                           'International Reputation','Weak Foot',]

# Decimales o dinero: 1.5, 2.4, $100
#Intercepcion sera continuo porque en FIF se calcula porque se toman muchos valores 
var_numericas_continuas = ['Potential','Value(£)','Wage(£)',
                           'Height(cm.)','Weight(lbs.)','Crossing','Finishing',
                          'HeadingAccuracy','ShortPassing','Volleys','Dribbling',
                          'Curve','FKAccuracy','LongPassing','BallControl','Acceleration',
                          'SprintSpeed','Agility','Reactions','Balance','ShotPower',
                          'Jumping','Stamina','Strength','LongShots','Aggression',
                          'Interceptions','Positioning','Vision','Penalties','Composure',
                          'Marking','StandingTackle','SlidingTackle','GKDiving','GKHandling',
                          'GKKicking','GKPositioning','GKReflexes','Release Clause(£)',
                          'DefensiveAwareness']

var_datetime = ['Joined','Contract Valid Until']
var_no_clasificadas = ['Photo','Flag','Club Logo']

print("Clasificacion de las variables\n")
print("Variables numericas discretas:", var_numericas_discretas)
print("Variables numericas continuas:", var_numericas_continuas)
print("Cantidad:", len(var_numericas_discretas)+len(var_numericas_continuas))
print(f"\nVariables categoricas ordinales:", var_categoricas_ordinales)
print("Variables categoricas nominales:", var_categoricas_nominales)
print("Cantidad:", len(var_categoricas_ordinales)+len(var_categoricas_nominales))
print(f"\nVariables no clasificadas:", var_no_clasificadas)
```

```python
top_nationalities = FIFA22_df['Nationality'].value_counts().nlargest(25)

plt.figure(figsize=(10, 10))
plt.pie(top_nationalities.values, labels=top_nationalities.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Nationalities Distribution')
plt.axis('equal')
plt.show()
```

```python
top_nationalities = FIFA22_df['Club'].value_counts().nlargest(25)

plt.figure(figsize=(10, 10))
plt.pie(top_nationalities.values, labels=top_nationalities.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Nationalities Distribution')
plt.axis('equal')
plt.show()
```

```python
# Get the top 10 nationalities and their counts
top_nationalities = FIFA22_df['Best Position'].value_counts().nlargest(20)

# Create a pie chart for top 10 nationalities
plt.figure(figsize=(10, 10))
plt.pie(top_nationalities.values, labels=top_nationalities.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Nationalities Distribution')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Age y Value(£)

Comparando La **edad** del jugador y su **valor en (£)** Creamos un gráfico de barras en el cual afirmamos que

La edad del jugador influye en su valor. Cuando el jugador es joven, su valor en el mercado es bajo, ya que su trayectoria es menor. A medida que envejece, su valor en (£) tiende a aumentar. Sin embargo, al superar los 34 años, su valor disminuye considerablemente, presumiblemente debido a la reducción de su capacidad física.

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='Age', y='Agility', data=FIFA22_df)
plt.title('Average Player Value (£) by Jersey Number')
plt.xlabel('Jersey Number')
plt.ylabel('Average Value (£)')
plt.xticks(rotation=90) 
plt.show()
```
